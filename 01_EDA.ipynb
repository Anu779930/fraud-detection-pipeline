**Dataset:** Credit Card Fraud Detection (Kaggle)  
**Records:** 284,807 transactions  


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

df = pd.read_csv('../data/sample_data.csv')
print(f'Dataset shape: {df.shape}')
df.head()
```

## 1. Dataset Overview

```python
print('=== Dataset Info ===')
print(df.info())
print('\n=== Summary Statistics ===')
df.describe()
```

## 2. Class Distribution (Fraud vs Non-Fraud)

```python
class_counts = df['Class'].value_counts()
fraud_pct = (class_counts[1] / len(df)) * 100

print(f'Non-Fraud (0): {class_counts[0]:,} ({100-fraud_pct:.2f}%)')
print(f'Fraud    (1): {class_counts[1]:,} ({fraud_pct:.2f}%)')

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Count plot
sns.countplot(x='Class', data=df, palette=['#2ecc71','#e74c3c'], ax=axes[0])
axes[0].set_title('Transaction Class Distribution', fontsize=14, fontweight='bold')
axes[0].set_xticklabels(['Non-Fraud', 'Fraud'])

# Pie chart
axes[1].pie(class_counts, labels=['Non-Fraud','Fraud'],
            colors=['#2ecc71','#e74c3c'], autopct='%1.2f%%', startangle=90)
axes[1].set_title('Fraud Percentage', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('../dashboard/class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3. Transaction Amount Analysis

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Amount distribution by class
df[df['Class']==0]['Amount'].hist(bins=50, alpha=0.6, color='#2ecc71',
                                   label='Non-Fraud', ax=axes[0])
df[df['Class']==1]['Amount'].hist(bins=50, alpha=0.6, color='#e74c3c',
                                   label='Fraud', ax=axes[0])
axes[0].set_title('Transaction Amount Distribution', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Amount ($)')
axes[0].legend()

# Boxplot
df.boxplot(column='Amount', by='Class', ax=axes[1],
           boxprops=dict(color='steelblue'))
axes[1].set_title('Amount by Class', fontsize=13, fontweight='bold')
axes[1].set_xticklabels(['Non-Fraud', 'Fraud'])

plt.tight_layout()
plt.savefig('../dashboard/amount_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

print('Non-Fraud — Mean Amount: $', round(df[df.Class==0].Amount.mean(), 2))
print('Fraud     — Mean Amount: $', round(df[df.Class==1].Amount.mean(), 2))
```

## 4. Correlation Heatmap

```python
plt.figure(figsize=(16, 10))
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, cmap='coolwarm', center=0,
            annot=False, linewidths=0.3)
plt.title('Feature Correlation Heatmap', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('../dashboard/correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Missing Values Check

```python
missing = df.isnull().sum()
print('Missing values per column:')
print(missing[missing > 0] if missing.any() else 'No missing values found ✓')
```
